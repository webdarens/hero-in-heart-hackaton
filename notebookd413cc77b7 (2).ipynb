```python
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Приводим тип объекта к нижнему регистру для единообразия
df['Тип.объекта_lower'] = df['Тип.объекта'].str.lower()

# Подвыборки
df_new = df[df['Тип.объекта_lower'] == 'новостройка']
df_sec = df[df['Тип.объекта_lower'] == 'вторичная']

# Проверка размеров
print("Подвыборки по типу объекта:")
print("Новостройки:", df_new.shape[0])
print("Вторичная недвижимость:", df_sec.shape[0])

# Формула для модели
formula = 'log_price_m2 ~ new_share_district + Площадь + Этаж + Возраст'

# Модели
model_new = smf.ols(formula, data=df_new).fit(cov_type='HC3')
model_sec = smf.ols(formula, data=df_sec).fit(cov_type='HC3')

# Таблица коэффициентов
coef_df = pd.DataFrame({
    'Коэффициент': [model_new.params['new_share_district'], model_sec.params['new_share_district']],
    '95% CI low': [model_new.conf_int().loc['new_share_district', 0], model_sec.conf_int().loc['new_share_district', 0]],
    '95% CI high': [model_new.conf_int().loc['new_share_district', 1], model_sec.conf_int().loc['new_share_district', 1]],
    'P-value': [model_new.pvalues['new_share_district'], model_sec.pvalues['new_share_district']]
}, index=['Новостройка', 'Вторичная'])

print("\n=== Эффект доли новостроек на цену (log_price_m2) ===")
print(coef_df)

# Визуализация
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='new_share_district', y='log_price_m2', hue='Тип.объекта_lower', alpha=0.6)
sns.regplot(data=df_new, x='new_share_district', y='log_price_m2', scatter=False, color='blue', label='fit новостройка')
sns.regplot(data=df_sec, x='new_share_district', y='log_price_m2', scatter=False, color='orange', label='fit вторичная')
plt.xlabel("Доля новостроек в районе (new_share_district)")
plt.ylabel("Лог цены за м² (log_price_m2)")
plt.title("Эффект доли новостроек на цену квартир")
plt.legend()
plt.show()
```

```python
# Только кластеры (не уникальные дома)
clustered = df[df['cluster'].notna()].copy()

# Доля новостроек в каждом кластере
cluster_new_share = clustered.groupby('cluster').apply(
    lambda x: (x['Тип.объекта'] == 'Новостройка').mean()
).rename('new_obj_share')

# Сколько объявлений в кластере
cluster_size = clustered.groupby('cluster').size().rename('n_listings')

cluster_info = pd.concat([cluster_new_share, cluster_size], axis=1)

# Смотрим распределение
print("Распределение доли новостроек в кластерах:")
print(cluster_info['new_obj_share'].describe())
print()

# Сколько кластеров при разных порогах
for threshold in [0.05, 0.10, 0.15, 0.20, 0.30, 0.50]:
    mixed = cluster_info[
        (cluster_info['new_obj_share'] >= threshold) & 
        (cluster_info['new_obj_share'] < 1.0)  # не чисто новостройки
    ]
    n_clusters = len(mixed)
    n_listings = clustered[clustered['cluster'].isin(mixed.index)].shape[0]
    n_vtorichka = clustered[
        (clustered['cluster'].isin(mixed.index)) & 
        (clustered['Тип.объекта'] == 'Вторичная')
    ].shape[0]
    print(f"Порог {threshold:.0%}: {n_clusters} кластеров, "
          f"{n_listings} объявлений, {n_vtorichka} вторичек → станут 'Вторичка_в_новостройке'")
```
